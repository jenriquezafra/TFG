```python
from src.stoc_system import StochasticSystem

r = np.array([1, 0.72, 1.53, 1.27])
a = np.array([
    [1, 1.09, 1.52, 0],
    [0, 1, 0.44, 1.36],
    [2.33, 0, 1, 0.47],
    [1.21, 0.51, 0.35, 1]
])
x0 = np.array([0.5, 0.5, 0.5, 0.5])
sigma = np.array([1.1, 1.1, 0.1, 0.1])
sigma = 0*sigma

N_pasos = 1e5
system = StochasticSystem(r, a, x0, 5000, 5000/N_pasos, sigma)
_, X = system.euler_maruyama()
```

```python
plt.plot(_, X)
```

```python
idx_check = int(0.95 * N_pasos)
row_check = X[idx_check, :]
n_ext = (row_check == 0).sum()
print(f"Number of extinct species: {n_ext} / {len(row_check)}")
```
